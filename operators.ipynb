```python
# List operations
A = [15, 20, 25]
B = [3, 6, 9]

print("A + B =", [A[0]+B[0], A[1]+B[1], A[2]+B[2]])
print("A - B =", [A[0]-B[0], A[1]-B[1], A[2]-B[2]])
print("A * B =", [A[0]*B[0], A[1]*B[1], A[2]*B[2]])
print("A / B =", [A[0]/B[0], A[1]/B[1], A[2]/B[2]])
print("A // B =", [A[0]//B[0], A[1]//B[1], A[2]//B[2]])
print("A % B =", [A[0]%B[0], A[1]%B[1], A[2]%B[2]])
print("A ** B =", [A[0]**B[0], A[1]**B[1], A[2]**B[2]])
print("A & B =", [A[0]&B[0], A[1]&B[1], A[2]&B[2]])
print("A | B =", [A[0]|B[0], A[1]|B[1], A[2]|B[2]])
print("A ^ B =", [A[0]^B[0], A[1]^B[1], A[2]^B[2]])

print("\n----------------------\n")

# For loop example
numbers = [1, 2, 3, 4, 5]
for n in numbers:
    print("numbers:", n)

print("\n----------------------\n")

# For loop with list
Courses = ["Java", "Python", "AIML"]
for C in Courses:
    print("Courses:", C)

print("\n----------------------\n")

# While loop example
i = 1
while i <= 5:
    print("i =", i)
    i += 1

print("\n----------------------\n")

# Reverse counting using while loop
x = 5
while x > 0:
    print("count:", x)
    x -= 1
     
A + B = [18, 26, 34]
A - B = [12, 14, 16]
A * B = [45, 120, 225]
A / B = [5.0, 3.3333333333333335, 2.7777777777777777]
A // B = [5, 3, 2]
A % B = [0, 2, 7]
A ** B = [3375, 64000000, 3814697265625]
A & B = [3, 4, 9]
A | B = [15, 22, 25]
A ^ B = [12, 18, 16]

----------------------

numbers: 1
numbers: 2
numbers: 3
numbers: 4
numbers: 5

----------------------

Courses: Java
Courses: Python
Courses: AIML

----------------------

i = 1
i = 2
i = 3
i = 4
i = 5

----------------------

count: 5
count: 4
count: 3
count: 2
count: 1
```
